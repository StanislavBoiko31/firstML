# 1. 

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
```

```python
digits = load_digits()
```

```python
print(digits.target[::100])
print(digits.data.shape)
print(digits.target.shape)
```

```python
figure, axes = plt.subplots(nrows = 4, ncols = 6, figsize=(8,8))

for item in zip(axes.ravel(), digits.images, digits.target):

    axes, image, target = item
    axes.imshow(image, cmap=plt.cm.gray_r)
    axes.set_xticks([])
    axes.set_yticks([])

    axes.set_title(target)
plt.tight_layout()
```

```python
figure, axes = plt.subplots(nrows = 6, ncols = 6, figsize=(8,8))

for item in zip(axes.ravel(), digits.images, digits.target):

    axes, image, target = item
    axes.imshow(image, cmap=plt.cm.gray_r)
    axes.set_xticks([])
    axes.set_yticks([])

    axes.set_title(target)
plt.tight_layout()
```

# 2.

```python
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, random_state=11, test_size=0.20)

print(f"Train shape: {X_train.shape}")

print(f"Test shape: {X_test.shape}")
```

# 3.

```python
knn = KNeighborsClassifier()
knn.fit(X=X_train, y=y_train)
```

# 4.

```python
predicted = knn.predict(X=X_test)
expected = y_test
```

# 5.

```python
print(f"Predicted: {predicted[:20]}")
print(f"Expected:  {expected[:20]}")
```

# 6.

## 6.1

```python
print(f"Score: {knn.score(X_test, y_test):.2%}")
```

## 6.2

```python
confusion = confusion_matrix(y_true=expected, y_pred=predicted)
print(confusion)
```

# 7.

```python
names = [str(digit) for digit in digits.target_names]
print(classification_report(expected, predicted, target_names=names))
```

# 8.

```python
svc = SVC(kernel='linear')
svc.fit(X=X_train, y=y_train)

predicted_svc = svc.predict(X=X_test)
expected_svc = y_test

print(f"Predicted: {predicted_svc[:20]}")
print(f"Expected:  {expected_svc[:20]}")

print(f"Score: {svc.score(X_test, y_test):.2%}")

confusion_svc = confusion_matrix(y_true=expected_svc, y_pred=predicted_svc)
print(confusion)

names_svc = [str(digit) for digit in digits.target_names]
print(classification_report(expected_svc, predicted_svc, target_names=names))

print()
print('GaussianNB')
print()
gaussianNB = GaussianNB()
gaussianNB.fit(X=X_train, y=y_train)

predicted_gaussianNB = gaussianNB.predict(X=X_test)
expected_gaussianNB = y_test

print(f"Predicted: {predicted_gaussianNB[:20]}")
print(f"Expected:  {expected_gaussianNB[:20]}")

print(f"Score: {gaussianNB.score(X_test, y_test):.2%}")

confusion_gaussianNB = confusion_matrix(y_true=expected_gaussianNB, y_pred=predicted_gaussianNB)
print(confusion)

names_gaussianNB = [str(digit) for digit in digits.target_names]
print(classification_report(expected_gaussianNB, predicted_gaussianNB, target_names=names))
```

# 9.

```python
for neighbors in range(1, 16):
    knn = KNeighborsClassifier(n_neighbors=neighbors).fit(X_train, y_train)
    accuracy = knn.score(X_test, y_test)
    
    print(f"Neighbors: {neighbors}, Accuracy: {accuracy:.2%}")
```
