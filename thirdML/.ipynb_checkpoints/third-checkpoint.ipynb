# 1. Порівняння з минулими результатами

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from matplotlib.colors import ListedColormap
```

```python
df = pd.read_csv('data2.csv')
df.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(df.Date.values.reshape(-1,1), df.Temperature.values.reshape(-1,1), random_state=11)
```

```python
print(f"Train: {x_train.shape}")
print(f"Test: {y_test.shape}")
```

```python
linear_regression = LinearRegression()

linear_regression.fit(X=x_train, y=y_train)

print(f"Coefficient: {linear_regression.coef_}")
print(f"Intercept: {linear_regression.intercept_}")
```

```python
predicted = linear_regression.predict(x_test).flatten()
expected = y_test.flatten()


for p, e in zip(predicted[::5], expected[::5]):
    print(f"predicted: {p:.2f}, expected: {e:.2f}")
```

```python
predict = (lambda x: linear_regression.coef_[0] * x + linear_regression.intercept_)

print(f"2019: {predict(2019)[0]}")
print(f"1890: {predict(1890)[0]}")
```

# 2. Згенеруйте набір даних та класифікуйте його використавши класифікатор SVC

```python
# Встановлюємо фіксоване значення seed для відтворюваності результатів
np.random.seed(3)
# Генеруємо випадкові дані 
X_xor = np.random.randn(200, 2)  # 200 точок з двох вимірів
y_xor = np.logical_xor(X_xor[:, 0] > 0, X_xor[:, 1] > 0)  
y_xor = np.where(y_xor, 1, -1)  

# Візуалізація початкових даних 
plt.scatter(X_xor[y_xor == 1, 0], X_xor[y_xor == 1, 1], c='b', marker='x', label='1')
plt.scatter(X_xor[y_xor == -1, 0], X_xor[y_xor == -1, 1], c='r', marker='s', label='-1')
plt.xlim([-3, 3])  # Встановлюємо межі для осі X
plt.ylim([-3, 3])  # Встановлюємо межі для осі Y
plt.legend(loc='best')  # Додаємо легенду
plt.show()

def plot_decision_regions(X, y, classifier, test_idx=None, resolution=0.02):
    markers = ('s', 'x', 'o', 'A', 'v')  # Символи для різних класів
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')  # Кольори для різних класів
    cmap = ListedColormap(colors[:len(np.unique(y))])  # Створюємо палітру кольорів для класів

    # Визначаємо межі для побудови сітки
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution), np.arange(x2_min, x2_max, resolution))

    # Отримуємо прогнози для всіх точок сітки
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)

    # Малюємо контурні ділянки рішень
    plt.contourf(xx1, xx2, Z, alpha=0.3, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())  # Межі для осі X
    plt.ylim(xx2.min(), xx2.max())  # Межі для осі Y

    # Виводимо точки навчальних даних
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], y=X[y == cl, 1], alpha=0.8, c=colors[idx], marker=markers[idx], label=cl)
    
    # Якщо є індекси тестового набору, вивести їх окремо
    if test_idx:
        X_test, y_test = X[test_idx, :], y[test_idx]
        plt.scatter(X_test[:, 0], X_test[:, 1], c='', edgecolor='black', alpha=1.0, linewidth=1, marker='o', s=100, label='Test set')

# Створення та навчання класифікатора SVС
svс = SVC(kernel='rbf', random_state=1, gamma=0.10, C=10.0)
svс.fit(X_xor, y_xor)

# Побудова графіку з областями рішень для навченої моделі
plot_decision_regions(X_xor, y_xor, classifier=svс)
plt.legend(loc='upper left')
plt.show()

```

# 3. KNeighborsClassifier, SVC та GaussianNB

```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Завантажуємо набір даних Wine
data = load_wine()
X = data.data
y = data.target

# Вибираємо 5 стовпців для масштабування
columns_to_scale = [0, 1, 2, 3, 4]  # alcohol, malic_acid, ash, alcalinity_of_ash, magnesium
scaler = StandardScaler()

# Масштабуємо обрані стовпці
X_scaled = X.copy()
X_scaled[:, columns_to_scale] = scaler.fit_transform(X[:, columns_to_scale])

# Розділення на навчальну та тестову вибірки
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, random_state=5)

# Ініціалізація класифікаторів
knn = KNeighborsClassifier(n_neighbors=5)
svc = SVC(kernel='rbf')
gnb = GaussianNB()

# Навчання моделей
knn.fit(X_train, y_train)
svc.fit(X_train, y_train)
gnb.fit(X_train, y_train)

# Прогнозування
y_pred_knn = knn.predict(X_test)
y_pred_svc = svc.predict(X_test)
y_pred_gnb = gnb.predict(X_test)

# Оцінка точності
accuracy_knn = accuracy_score(y_test, y_pred_knn)
accuracy_svc = accuracy_score(y_test, y_pred_svc)
accuracy_gnb = accuracy_score(y_test, y_pred_gnb)


# Виведення результатів
print("KNeighborsClassifier Accuracy: ", accuracy_knn)
print("SVC Accuracy: ", accuracy_svc)
print("GaussianNB Accuracy: ", accuracy_gnb)
```
